```python
class Solution:
    def numSub(self, s: str) -> int:
        total = 0
        count = 0

        for c in s:
            if c == '1':
                count += 1
                total += count
            else:
                count = 0

        return total % (10**9 + 7)

solution = Solution()
%time solution.numSub(s)
```
